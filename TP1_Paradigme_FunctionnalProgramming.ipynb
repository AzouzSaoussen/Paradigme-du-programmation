```python
L= [ 1, 5 , 7 , 9 , 15, 20, 50, 90, 9, 3, 12, 30]
from random import seed
from random import randint
seed(1)
for _ in range(50000):
	L.append(randint(0, 200))
print(L)
```

```python
import time
from functools import reduce 
tic = time.perf_counter()
sum = 0
def add(x, y):
	return x + y
sum = reduce(add, L)
toc = time.perf_counter()
print (sum, toc-tic)
```

```python
tic = time.perf_counter()
sum = 0
for x in L:
	sum += x
toc = time.perf_counter()
print (sum, toc-tic)
```

```python
def f(x,n):
  if n == 0:
    return 1
  if n % 2 == 0:
    p = f(x, n/2)
    return p*p
  else:
    return x*f(x,n-1)

f(2,20)
```

```python
def fac(n):
  if n < 0:
    print("Factoriel ne peut tre trouv pour les nombres ngatifs")
  else:
    if n== 0 or n == 1:
       return 1  
    else:
        return n* fac(n-1)
    
fac(-1)
```

```python
def sommeDiviseurs(n):
    somme = 0
    for i in range(1, n):
        if n%i == 0:
            somme = somme + i
    return somme
sommeDiviseurs(220)
```

```python
def nombres_amicaux(m , n):
  if m > 0 and n > 0 :
    if sommeDiviseurs(m ) == n and sommeDiviseurs(n) == m:
        return True
    else:
        return False
  else:
    print("les arguments saisis doivent etre deux entiers positifs")      
```

```python
nombres_amicaux(220, 284)
```
